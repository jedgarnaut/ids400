```python
#then here is the multiple models with confusion matrices and cross validated adapted from the lecture for assignment 7

#The SVM takes a while to run.

#I'm a bit curious about the logistic and SVM outputs.


# multiple models

import sklearn
from sklearn.svm import LinearSVC
import warnings
warnings.filterwarnings('ignore')
from sklearn.metrics import confusion_matrix
from sklearn.svm import SVC
from sklearn.neighbors import KNeighborsClassifier

X_train, X_test, y_train, y_test = train_test_split(arrayX, arrayY, test_size=0.1, random_state = 42)


models = []
models.append(("LR", LogisticRegression(solver = "lbfgs", max_iter = 100000)))
models.append(("KNN", sklearn.neighbors.KNeighborsClassifier(n_neighbors=5)))     
models.append(("CART", DecisionTreeClassifier()))
models.append(("Linear", LinearSVC()))
models.append(("SVM", SVC()))
               
#model evaluation in turn

results = []
names = []
for name, model in models:
    kfold = model_selection.KFold(n_splits = 10, shuffle = False)
    cv_results = model_selection.cross_val_score(model, X_train, y_train, cv=kfold, scoring = "accuracy")
    model.fit(X_train, y_train)
    y_pred=model.predict(X_test)
    cm = confusion_matrix(y_test, y_pred)
    results.append(cv_results)
    names.append(name)
    msg = "%s: %f (%f)" % (name, cv_results.mean(), cv_results.std())
    print(msg)
    print(cm)
```

```python
# IQR is the difference between the 75th percentile (Q3) and the 25th percentile (Q1).
# Outliers are then defined as observations that are below [Q1 - 1.5*IQR] and are above [Q3 + 1.5*IQR]

Q1 = df['Education'].quantile(0.25)
Q3 = df['Education'].quantile(0.75)

IQR = Q3 - Q1

# Define outliers
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# Filter out outliers
filtered_df = df[(df['Education'] >= lower_bound) & (df['Education'] <= upper_bound)]

filtered_df.shape
```
